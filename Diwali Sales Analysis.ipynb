# Exploratory Data Analysis

#### Gender

```python
ax = sns.countplot( x = 'Gender', data = data)
for bars in ax.containers:
    ax.bar_label(bars)
```

*This chart is showing that Females are doing more shopping than mens*

```python
sales_gen = data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending=False)
sns.barplot(x= 'Gender',y='Amount',data=sales_gen);
```

*Again it is not surprising that females are buing more compare to mens , on the bases of amount also.*

## Age

```python
ax=sns.countplot(data = data, x='Age Group' , hue ='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

*This chart shows that females are buying more in all age groups , specially from 18-36 .*

```python
sales_age = data.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y= 'Amount',data=sales_age);
```

*Again it is not surprising that females are buing more compare to mens , on the bases of amount also.*

## State

```python
sales_state = data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data=sales_state,x = 'State',y='Orders');
for bars in ax.containers:
    ax.bar_label(bars);
```

*This chart clearly shows that we are mostly recieving orders from Uttar Pradesh , Maharashtra, Karnataka.*

```python
sales_state = data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x= 'State',y= 'Amount');
```

*Again it is  clearly showing that we are mostly recieving orders from Uttar Pradesh , Maharashtra, Karnataka , on the bases of Amount.*

## Marital Status

```python
ax = sns.countplot(data = data, x = 'Marital_Status' , hue = 'Gender')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

*Here 0 represents Unmarried and 1 represents Married , More Females are Married compare to Males.* 

```python
sales_state = data.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender');
```

*This chart shows that married females and unmarried females are more buying products compare to married males and unmarried males.*

## Product Category

```python
sns.set(rc={'figure.figsize':(20,15)})
ax = sns.countplot(data = data, y = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

*People are more likely to buy - clothing & apparel , food , electronics & gadgets.*

```python
sales_state = data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

*People are more spending on food compare to other products on the bases of amount.* 

```python
sales_state = data.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders');
```

*This chart shows that which product ID is recieving more others than others.*
